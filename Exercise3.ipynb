```python
#Exercise 3
#Q1
#Draw a line in a diagram from position (1,3) to (2,10) then to (6,12) and finally to position (18,20). Mark each point 
#with a beautiful green colour and set line colour to red and line style dotted

import numpy as np
import matplotlib.pyplot as plt
x=(1,6,18)
y=(3,12,20)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,'--',color='red',marker='o',markersize=10,label='graph',markerfacecolor='green',markeredgecolor='green')
plt.grid(True)
plt.show()
```

```python
#Plot a graph using table
#Q2

import numpy as np
import matplotlib.pyplot as plt
x=(12,14,16,18,20,22,24)
y=(100,200,250,400,300,450,500)
plt.title("Graph")
plt.xlabel("Temperature in C")
plt.ylabel("Scales")
plt.plot(x,y,color='orange',marker='o',markersize=10,label='graph',markerfacecolor='yellow',markeredgecolor='yellow')
plt.grid(True)
plt.show()
```

```python
#Q3
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[2,3,5,7,11]
plt.plot(x,y1,label='Line1')
plt.plot(x,y2,label='Line2')
plt.title('Multiple lines on same plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
#Q5
#Write a python program to create multiple plots
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[10,20,30,40,50]
x2=[1,2,3,4,5]
y2=[50,40,30,20,10]
fig,axs=plt.subplots(1,2,figsize=(10,5))

axs[0].plot(x1,y1,color='blue',marker='o',label='Dataset 1')
axs[0].set_title('Plot 1: Dataset 1')
axs[0].set_xlabel('X Axis')
axs[0].set_ylabel('Y Axis')
axs[0].legend()

axs[1].plot(x2,y2,color='blue',marker='o',label='Dataset 2')
axs[1].set_title('Plot 2: Dataset 2')
axs[1].set_xlabel('X Axis')
axs[1].legend()

plt.tight_layout()
plt.show()
```
