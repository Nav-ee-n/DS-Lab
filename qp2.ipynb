```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,11)
y=3*(x)+5
plt.title("Graph")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y,color='orange',marker='o',markersize=12,label='graph')
plt.legend()
plt.show()
```
