```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("Record")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.title("Record")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y,"g")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title("Record")

plt.subplot(2,1,1)
plt.plot(x,z)
plt.title("Data")
plt.show()
```
